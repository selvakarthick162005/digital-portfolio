Using matplot log any diagram spykit search mean square error

```

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 1: Generate simple linear data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

# Step 2: Train Linear Regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Step 3: Calculate Mean Squared Error
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", round(mse, 2))

# Step 4: Plot data, predictions, and log scale
plt.figure(figsize=(8, 5))
plt.plot(X, y, 'bo-', label='Actual')         # Blue dots: actual data
plt.plot(X, y_pred, 'r--', label='Predicted')  # Red dashed: predicted
plt.yscale('log')                              # Apply log scale to Y-axis
plt.xlabel('X')
plt.ylabel('Y (Log Scale)')
plt.title('Linear Regression with Log Plot')
plt.legend()
plt.grid(True)
plt.show()
```
